```python
import sys
```

```python
def travel_grid(m,n):
    dp = [[None for _ in range(n)] for _ in range(m)]
    dp[0][0] = 1
    for i in range(m):
        for j in range(n):
            if i ==0 and j ==0:
                continue
            if i == 0:
                dp[i][j] = dp[i][j-1]
            elif j == 0:
                dp[i][j] = dp[i-1][j]
            else:
                dp[i][j] = dp[i-1][j] + dp[i][j-1]
    return dp[m-1][n-1]
```

```python
travel_grid(5,5)
```

```python
def travel_grid(m, n):
    dp = [[1] * n for _ in range(m)]
    
    for i in range(1, m):
        for j in range(1, n):
            dp[i][j] = dp[i - 1][j] + dp[i][j - 1]
    
    print("Size: ", sys.getsizeof(dp), "bytes")
    return dp[-1][-1]
```

```python
travel_grid(5000,5000)
```

```python
def travel_grid_opt(m,n):

    if n > m:
        m,n = n,m
    
    dp = [[1 for _ in range(n)] for _ in range(2)]   # Space optimization

    for i in range(1,m):
        for j in range(1,n):
            dp[1][j] = dp[1][j-1] + dp[0][j]
        dp[0] = dp[1][:]
    
    print("Size: ", sys.getsizeof(dp), "bytes")
    return dp[-1][-1]
```

```python
travel_grid_opt(5000,5000)
```

```python
def travel_grid_dict(m,n, dp = {}):
    key = str(m) + "," + str(n)

    if key in dp:
        return dp[key], sys.getsizeof(dp)

    if m == 1 or n == 1 : return 1, 0

    dp[key] = travel_grid_dict(m-1,n,dp)[0] + travel_grid_dict(m,n-1,dp)[0]

    return dp[key], sys.getsizeof(dp)
```

```python
travel_grid_dict(50,50)  # size of 51 968 bytes
```
